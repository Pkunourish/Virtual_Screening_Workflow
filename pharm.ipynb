```python
import pandas as pd
```

```python
df_map = pd.read_csv('./merge_1443.csv')
rmsd = pd.read_csv('./1443_rmsd_select.csv')
rmsd
```

```python
sort_rmsd = rmsd.merge(
    df_map[['standardized_smiles', 'score_vina', 'score_glide']].copy(), 
    on = 'standardized_smiles',
    how='left',
    suffixes=('', '_dup')
)
sort_rmsd
#sort_rmsd.sort_values(by = 'rmsd').copy().to_csv('./sort_1443_rmsd.csv', index=False)
```

```python
pharm = pd.read_csv('./sort_1443_PharmFit.csv')
#pharm.to_csv('./sort_1443_PharmFit.csv', index=False)
pharm
```

```python
sort_pharm = sort_rmsd.merge(
    pharm[['vina_title', 'PharmFit_score']].copy(), 
    on = 'vina_title',
    how='left',
    suffixes=('', '_dup')
)
final_pharm = sort_pharm.sort_values(by = 'PharmFit_score', ascending=False).copy()
final_pharm['PharmFit_score'] = pd.to_numeric(final_pharm['PharmFit_score'], errors='coerce')
final_pharm[final_pharm['PharmFit_score']>=0.50].copy().to_csv('./1443_pharm_select.csv', index=False)
```

```python
final_pharm[final_pharm['PharmFit_score']>=0.50]['vina_title'].copy().to_csv('./1443_pharm_vinalist.txt', index=False, header=False)
final_pharm[final_pharm['PharmFit_score']>=0.50]['glide_title'].copy().to_csv('./1443_pharm_glidelist.txt', index=False, header=False)
```
